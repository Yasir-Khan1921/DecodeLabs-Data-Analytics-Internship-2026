## 1. Introduction

### What is Exploratory Data Analysis (EDA)?
**Exploratory Data Analysis (EDA)** is a critical stage in the data science pipeline that involves investigating, summarizing, and visualizing datasets to understand their main characteristics. First introduced by the American mathematician John Tukey, EDA is designed to help analysts look at data before making any assumptions or building complex predictive models. It relies on a combination of descriptive statistics and visual methods (like histograms, scatter plots, and box plots) to explore data distribution, check for anomalies/outliers, verify data quality, and discover interesting relationships.

### Objective of the Project
The goal of this project is to conduct a complete, professional, and portfolio-ready Exploratory Data Analysis (EDA) on an e-commerce transactional dataset. This involves:
- **Loading and Inspecting**: Understanding data shape, schema, and basic column meanings.
- **Quality Assessment**: Identifying missing data and duplicate entries, and applying remedies.
- **Descriptive Summary**: Calculating statistical attributes (mean, median, dispersion, quartiles).
- **Distribution Analysis**: Visualizing distribution shapes and computing skewness indices.
- **Outlier Identification**: Isolating anomalous entries using the Interquartile Range (IQR) method.
- **Correlation and Co-dependencies**: Quantifying variables correlation and plotting heatmaps.
- **Bivariate Relationship Analysis**: Reviewing how target metrics (e.g. Total Price) depend on product cost and quantity.
- **Strategic Recommendations**: Converting findings into actionable business insights.

## 5. Missing Values Analysis

We check for missing/null values using `isnull().sum()`. To ensure complete visibility, we will also create a Seaborn Heatmap of null values. If any missing values are found, we handle them using appropriate techniques (mean for normal numerical data, median for skewed numerical data, and mode for categorical data).

```python
# ============================================================
# Step 5a: Identify missing values using isnull().sum()
# ============================================================
null_summary = df.isnull().sum()
print("═" * 50)
print("MISSING VALUES ANALYSIS")
print("═" * 50)
for column, count in null_summary.items():
    percentage = (count / len(df)) * 100
    print(f"  {column:<20s} | Nulls: {count:<4d} | Percentage: {percentage:.2f}%")

total_null_count = null_summary.sum()
print(f"\n📌 Total missing fields in dataset: {total_null_count}")
```

```python
# ============================================================
# Step 5b: Visualize missing values using a Seaborn Heatmap
# ============================================================
plt.figure(figsize=(10, 4))
# Plot heatmap highlighting any missing/null values in yellow
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap="viridis")
plt.title("Visual Heatmap of Missing Values (Solid color indicates zero nulls)", pad=15)
plt.show()
```

```python
# ============================================================
# Step 5c: Handle missing values (if any exist)
# ============================================================
if total_null_count > 0:
    print("🔧 Missing values detected. Applying appropriate imputation techniques...")
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            if df[col].dtype in ["int64", "float64"]:
                # Check skewness to decide between Mean and Median
                skewness = df[col].skew()
                if abs(skewness) > 1.0:
                    # Skewed data -> Use Median
                    fill_val = df[col].median()
                    method = f"Median (Skewness: {skewness:.2f})"
                else:
                    # Symmetrical data -> Use Mean
                    fill_val = df[col].mean()
                    method = f"Mean (Skewness: {skewness:.2f})"
                df[col] = df[col].fillna(fill_val)
            else:
                # Categorical data -> Use Mode
                fill_val = df[col].mode()[0]
                method = "Mode (Categorical)"
            print(f"   Filled nulls in {col} using {method}: value = {fill_val}")
    print(f"✅ Imputation completed. Current null count: {df.isnull().sum().sum()}")
else:
    print("✅ No missing values exist. The dataset is fully populated and clean.")
```

## 7. Distribution Analysis

We visualize the shape of the data using **Histograms** (binned frequencies) and **KDE (Kernel Density Estimate)** curves. We also calculate skewness to classify the distribution:
- **Symmetrical**: Data is distributed evenly around the center. Mean $\approx$ Median. Skewness index is close to 0 (typically between $-0.5$ and $+0.5$).
- **Right-skewed (Positive Skew)**: The tail of the distribution extends to the right. Mean $>$ Median. Skewness index $> 0.5$.
- **Left-skewed (Negative Skew)**: The tail of the distribution extends to the left. Mean $<$ Median. Skewness index $< -0.5$.

```python
# ============================================================
# Step 7: Plot distribution curves and evaluate skewness
# ============================================================
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for idx, col in enumerate(numerical_cols):
    # Calculate skewness index
    skew_val = df[col].skew()
    
    # Plot Histogram combined with Kernel Density Estimate (KDE) curve
    sns.histplot(df[col], kde=True, ax=axes[idx], color="teal", stat="density", alpha=0.5)
    
    # Add vertical reference lines for Mean and Median to show skewness
    mean_val = df[col].mean()
    median_val = df[col].median()
    axes[idx].axvline(mean_val, color="red", linestyle="--", linewidth=1.5, label=f"Mean: {mean_val:.2f}")
    axes[idx].axvline(median_val, color="blue", linestyle="-.", linewidth=1.5, label=f"Median: {median_val:.2f}")
    
    # Configure plot labels
    axes[idx].set_title(f"Distribution of {col}\nSkewness Index: {skew_val:.4f}")
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel("Density")
    axes[idx].legend(loc="upper right")

plt.suptitle("Distribution Curves (Histogram + KDE) of E-Commerce Variables", y=1.02, fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()

# print descriptive labels of distributions
print("═" * 70)
print("SKEWNESS CLASSIFICATION REPORT")
print("═" * 70)
for col in numerical_cols:
    val = df[col].skew()
    if abs(val) < 0.1:
        desc = "Symmetrical (Normal-like)"
    elif val >= 0.1 and val <= 0.5:
        desc = "Slightly Right-Skewed"
    elif val > 0.5:
        desc = "Moderately/Highly Right-Skewed (Positive Skew)"
    elif val <= -0.1 and val >= -0.5:
        desc = "Slightly Left-Skewed"
    else:
        desc = "Moderately/Highly Left-Skewed (Negative Skew)"
    print(f"  {col:<15s} | Skewness: {val:>7.4f} | Distribution Type: {desc}")
```

## 8. Outlier Detection

An **outlier** is an observation that lies an abnormal distance from other values in the dataset. To identify outliers objectively, we use the **Interquartile Range (IQR) method**:

$$\text{IQR} = Q3 - Q1$$
$$\text{Lower Bound} = Q1 - 1.5 \times \text{IQR}$$
$$\text{Upper Bound} = Q3 + 1.5 \times \text{IQR}$$

Any data point falling below the Lower Bound or above the Upper Bound is flagged as an outlier. We will visualize these points using **Boxplots**.

```python
# ============================================================
# Step 8a: Create Boxplots for numerical columns
# ============================================================
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for idx, col in enumerate(numerical_cols):
    # Draw boxplot showing whiskers and outliers (diamonds)
    sns.boxplot(x=df[col], ax=axes[idx], color="coral", flierprops=dict(markerfacecolor="red", marker="D", markersize=6))
    axes[idx].set_title(f"Boxplot of {col}")
    axes[idx].set_xlabel(col)

plt.suptitle("Boxplots for Numerical Columns (Highlighting Outlier Diamonds)", y=1.02, fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# Step 8b: Calculate bounds and isolate outliers using IQR formula
# ============================================================
print("═" * 80)
print("OUTLIER DETECTED VIA IQR METHOD")
print("═" * 80)

for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Isolate outlier rows
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    print(f"📊 Column: {col}")
    print(f"   Bounds: Lower Limit = {lower_bound:.2f} | Upper Limit = {upper_bound:.2f}")
    print(f"   Outlier Count: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
    if len(outliers) > 0:
        sample_vals = sorted(outliers[col].unique())
        print(f"   Top Outlier Values: {sample_vals[:10]}")
    print("-" * 80)
```

## 9. Correlation Analysis

Correlation analysis measures the linear relationship between variables using Pearson's correlation coefficient ($r$, ranging from $-1$ to $+1$):

- **Positive Correlation ($r > 0$)**: As variable A increases, variable B increases. If A goes up, B goes up.
- **Negative Correlation ($r < 0$)**: As variable A increases, variable B decreases. If A goes up, B goes down.
- **No Correlation ($r \approx 0$)**: There is no linear pattern or directional association between variables.

We calculate the correlation matrix for our numerical columns and plot a heatmap using Seaborn.

```python
# ============================================================
# Step 9: Compute Pearson correlation matrix and draw heatmap
# ============================================================
# Calculate Pearson correlation matrix
corr_mat = df[numerical_cols].corr()

print("═" * 60)
print("PEARSON CORRELATION COEFFICIENTS")
print("═" * 60)
print(corr_mat.round(4))
print() # print a line space

# Plot correlation matrix using Seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_mat, annot=True, cmap="coolwarm", fmt=".4f", linewidths=1.5, vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Numerical Features", pad=20, fontsize=14, fontweight="bold")
plt.show()
```

## 10. Relationship Analysis

We explore interactions between variables using **Scatter Plots** (to analyze bivariate relationships) and a **Pairplot** (to inspect all variables simultaneously).

```python
# ============================================================
# Step 10a: Create Scatter plots for TotalPrice vs UnitPrice & Quantity
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Scatter plot: UnitPrice vs TotalPrice (colored/sized by Quantity)
sns.scatterplot(data=df, x="UnitPrice", y="TotalPrice", hue="Quantity", size="Quantity",
                sizes=(20, 150), palette="viridis", alpha=0.8, ax=axes[0])
axes[0].set_title("Total Order Price vs Product Unit Price\n(Grouped by Quantity)")
axes[0].set_xlabel("Unit Price ($)")
axes[0].set_ylabel("Total Order Value ($)")

# Strip plot: Quantity vs TotalPrice (categorized by Product)
sns.stripplot(data=df, x="Quantity", y="TotalPrice", hue="Product", jitter=0.25,
              palette="tab10", alpha=0.7, ax=axes[1])
axes[1].set_title("Total Order Price Distribution by Quantity\n(Categorized by Product)")
axes[1].set_xlabel("Quantity Purchased")
axes[1].set_ylabel("Total Order Value ($)")
axes[1].legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Product")

plt.tight_layout()
plt.show()
```
