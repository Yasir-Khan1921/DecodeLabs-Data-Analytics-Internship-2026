## Step 4: Identify Missing / Null Values

Use `isnull().sum()` to detect missing values in each column. This is critical to determine which columns need imputation.

```python
# ============================================================
# Step 4: Identify all missing/null values using isnull().sum()
# ============================================================
missing_values = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df) * 100).round(2)

# Create a summary DataFrame for missing values
missing_df = pd.DataFrame({
    "Column": missing_values.index,
    "Missing Count": missing_values.values,
    "Missing %": missing_percent.values
})

# Sort by missing count descending
missing_df = missing_df.sort_values("Missing Count", ascending=False).reset_index(drop=True)

print("═" * 60)
print("MISSING VALUES ANALYSIS")
print("═" * 60)
print(missing_df.to_string(index=False))
print(f"\n🔍 Total missing values: {df.isnull().sum().sum()}")
print(f"📌 Columns with missing data: {(missing_values > 0).sum()}")
```

## Step 5: Handle Missing Values

### Strategy:
- **CouponCode** (Categorical): Fill with **Mode** (most frequent value)
  - Mode is the appropriate technique for categorical data as mean/median are not applicable to non-numeric values.

> **Note:** All numerical columns (Quantity, UnitPrice, ItemsInCart, TotalPrice) have zero missing values, so mean/median imputation is not required. However, we demonstrate the approach below for completeness.

```python
# ============================================================
# Step 5: Handle missing values using appropriate techniques
# ============================================================

# --- Technique Demonstration for Numerical Columns ---
# Although no numerical columns have missing values in this dataset,
# we demonstrate the techniques that WOULD be used:

numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns
print("═" * 60)
print("NUMERICAL COLUMNS - MISSING VALUE CHECK")
print("═" * 60)
for col in numerical_cols:
    skewness = df[col].skew()
    technique = "Median (skewed)" if abs(skewness) > 1 else "Mean (normal)"
    print(f"  {col:<15s} | Missing: {df[col].isnull().sum():<5d} | Skewness: {skewness:.2f} | Recommended: {technique}")

print("\n✅ No numerical columns have missing values — no imputation needed.")
print("   If they did: Mean for normally distributed, Median for skewed data.\n")

# --- Handle Categorical Missing Values ---
print("═" * 60)
print("CATEGORICAL COLUMNS - MISSING VALUE HANDLING")
print("═" * 60)

# CouponCode: Fill with Mode (most frequent value)
coupon_mode = df["CouponCode"].mode()[0]
print(f"\n  CouponCode:")
print(f"    Missing values: {df['CouponCode'].isnull().sum()}")
print(f"    Mode (most frequent): '{coupon_mode}'")

df["CouponCode"] = df["CouponCode"].fillna(coupon_mode)

print(f"    After fillna: {df['CouponCode'].isnull().sum()} missing values")
print(f"\n✅ Missing values handled successfully using Mode for categorical data!")
```

## Step 6: Identify Duplicate Rows

Use `duplicated().sum()` to check for any duplicate records in the dataset.

```python
# ============================================================
# Step 6: Identify duplicate rows using duplicated().sum()
# ============================================================
duplicate_count = df.duplicated().sum()

print("═" * 60)
print("DUPLICATE ROWS ANALYSIS")
print("═" * 60)
print(f"  Total duplicate rows found: {duplicate_count}")
print(f"  Total rows in dataset: {len(df)}")
print(f"  Percentage of duplicates: {(duplicate_count/len(df)*100):.2f}%")

if duplicate_count == 0:
    print("\n✅ No duplicate rows detected in the dataset.")
else:
    print(f"\n⚠️  {duplicate_count} duplicate rows detected — will be removed in next step.")
```

## Step 7: Remove Duplicate Rows

Use `drop_duplicates()` to remove any duplicate records and ensure data integrity.

```python
# ============================================================
# Step 7: Remove duplicate rows using drop_duplicates()
# ============================================================
rows_before = len(df)
df = df.drop_duplicates()
rows_after = len(df)
rows_removed = rows_before - rows_after

print("═" * 60)
print("DUPLICATE REMOVAL RESULTS")
print("═" * 60)
print(f"  Rows before: {rows_before}")
print(f"  Rows after:  {rows_after}")
print(f"  Rows removed: {rows_removed}")
print(f"\n✅ drop_duplicates() executed — {rows_removed} duplicate(s) removed.")
print(f"   Dataset now has {rows_after} clean rows.")
```

## Step 8: Correct Data Formats

Ensure all data is properly formatted:
- ✅ Convert dates to **YYYY-MM-DD** format
- ✅ Remove **extra spaces** from text columns
- ✅ Standardize **text capitalization** (Title Case)
- ✅ Ensure **numerical columns** have correct data types

```python
# ============================================================
# Step 8a: Convert Date column to YYYY-MM-DD string format
# ============================================================
print("═" * 60)
print("DATE FORMAT CORRECTION")
print("═" * 60)
print(f"  Before: {df['Date'].dtype}")
print(f"  Sample (before): {df['Date'].iloc[0]}")

# Ensure Date is datetime then format to YYYY-MM-DD
df["Date"] = pd.to_datetime(df["Date"])
df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")

print(f"  After:  {df['Date'].dtype}")
print(f"  Sample (after):  {df['Date'].iloc[0]}")
print(f"\n✅ Dates converted to YYYY-MM-DD format.")
```

```python
# ============================================================
# Step 8b: Remove extra spaces from text columns
# ============================================================
print("═" * 60)
print("TEXT CLEANING - REMOVE EXTRA SPACES")
print("═" * 60)

text_columns = df.select_dtypes(include="object").columns
print(f"  Text columns identified: {list(text_columns)}\n")

for col in text_columns:
    # Count values with leading/trailing whitespace before cleaning
    space_count = df[col].apply(lambda x: x != x.strip() if isinstance(x, str) else False).sum()
    # Strip leading and trailing spaces
    df[col] = df[col].str.strip()
    # Also replace multiple internal spaces with single space
    df[col] = df[col].str.replace(r"\s+", " ", regex=True)
    print(f"  {col:<20s} | Spaces fixed: {space_count}")

print(f"\n✅ Extra spaces removed from all {len(text_columns)} text columns.")
```

```python
# ============================================================
# Step 8c: Standardize text capitalization (Title Case)
# ============================================================
print("═" * 60)
print("TEXT CAPITALIZATION STANDARDIZATION")
print("═" * 60)

# Apply Title Case to selected categorical columns
capitalize_cols = ["Product", "PaymentMethod", "OrderStatus", "ReferralSource", "CouponCode"]

for col in capitalize_cols:
    df[col] = df[col].str.title()
    sample_vals = df[col].unique()[:3]
    print(f"  {col:<20s} | Sample values: {list(sample_vals)}")

print(f"\n✅ Text capitalization standardized to Title Case.")
```

```python
# ============================================================
# Step 8d: Ensure numerical columns have correct data types
# ============================================================
print("═" * 60)
print("NUMERICAL DATA TYPE VERIFICATION")
print("═" * 60)

# Verify and fix numerical columns
numerical_checks = {
    "Quantity": "int64",
    "UnitPrice": "float64",
    "ItemsInCart": "int64",
    "TotalPrice": "float64"
}

for col, expected_type in numerical_checks.items():
    current_type = str(df[col].dtype)
    if current_type != expected_type:
        df[col] = df[col].astype(expected_type)
        status = "🔧 Converted"
    else:
        status = "✅ Correct"
    print(f"  {col:<15s} | Expected: {expected_type:<10s} | Current: {current_type:<10s} | {status}")

print(f"\n✅ All numerical columns verified with correct data types.")
```

Verify that all cleaning operations were successful:

- ✅ Zero duplicate rows
- ✅ Dates correctly formatted (YYYY-MM-DD)
- ✅ All data types correct

```python
# ============================================================
# Step 9: Comprehensive Final Verification
# ============================================================
print("╔" + "═" * 58 + "╗")
print("║" + " FINAL DATA QUALITY VERIFICATION REPORT ".center(58) + "║")
print("╠" + "═" * 58 + "╣")

# Check 1: Missing Values
missing_total = df.isnull().sum().sum()
status_1 = "✅ PASS" if missing_total == 0 else "❌ FAIL"
print(f"║  1. Missing Values:    {missing_total:<10d}          {status_1:<10s}  ║")

# Check 2: Duplicate Rows
dup_total = df.duplicated().sum()
status_2 = "✅ PASS" if dup_total == 0 else "❌ FAIL"
print(f"║  2. Duplicate Rows:    {dup_total:<10d}          {status_2:<10s}  ║")

# Check 3: Date Format
date_check = all(df["Date"].str.match(r"^\d{4}-\d{2}-\d{2}$"))
status_3 = "✅ PASS" if date_check else "❌ FAIL"
print(f"║  3. Date Format:       {'YYYY-MM-DD':<10s}          {status_3:<10s}  ║")

# Check 4: Data Types
type_check = (df["Quantity"].dtype == "int64" and 
              df["UnitPrice"].dtype == "float64" and
              df["TotalPrice"].dtype == "float64" and
              df["ItemsInCart"].dtype == "int64")
status_4 = "✅ PASS" if type_check else "❌ FAIL"
print(f"║  4. Data Types:        {'Correct':<10s}          {status_4:<10s}  ║")

print("╠" + "═" * 58 + "╣")

all_pass = all([missing_total == 0, dup_total == 0, date_check, type_check])
overall = "ALL CHECKS PASSED ✅" if all_pass else "SOME CHECKS FAILED ❌"
print(f"║  {'OVERALL:':<20s} {overall:<36s}  ║")
print("╚" + "═" * 58 + "╝")

print(f"\n📊 Final Dataset Shape: {df.shape[0]} rows × {df.shape[1]} columns")
```
